```python
add = lambda num1,num2 : num1+num2
add(3,4)
```

```python
even_odd = lambda num : "even" if num%2==0 else "odd"
even_odd(4)
```

```python
nums = [1,2,3,4,5,6,7,8]
list(map(lambda i : i**2,nums))
```

```python
nums = [1,2,3,4,5,6,7,8]
list(map(lambda i : "even" if i%2==0 else "odd",nums))
```

```python
nums = [1,2,3,4,5,6,8]
list(filter(lambda num : True if num%2==0 else False, nums))
```

```python
nums = [1,2,3,4,5,6,8]
list(filter(lambda num : True if num%2!=0 else False, nums))
```

```python
strings = ["good morning everyone","how are you all",
           "are you understanding lambda espresssions"]
list(map(lambda s: len(s.split()) ,strings))
```

```python
# Write a program to count 
# how many words in a sentence have length > 4 using lambda and filter
s = "Python lambda functions are useful and powerful"
words = s.split() #["python","lambda",]
words2= list(filter(lambda x: len(x)>4,words))
print(words2)
```
