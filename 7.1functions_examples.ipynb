```python
def second(a,b,c):
    if a > b and a > c:
        if b > c:
            print(b, 'is second largest')
        else:
            print(c, 'is second largest')
    elif b > c:
        if a > c:
            print(a, 'is second largest')
        else:
            print(c, 'is second largest')
    else:
        if a > b:
            print(a, 'is second largest')
        else:
            print(b, 'is second largest')
```

```python
second(4,6,8)
```

```python
def check_prime(num):
    if num<=1:
        return False
    else:
        for i in range(2,int(num**(1/2))+1):
            if num%i==0:
                return False
        else:
            return True
            
```

```python
list(filter(check_prime,range(1,101)))
```
