###  Given a list of numbers print the elemnets of list one by one using range function

```python
nums = [20,30,40,10,50]
for i in range(len(nums)):
    print(nums[i], end = '\n')
```

###  Given a list of numbers print the elemnets of list one by one by for each loop


```python
nums = [20,30,40,10,50]
for i in nums:
    print(i, end = '\n')
```

### Given a list of numbers take a number as input from the user and print 'the number is present' if the number is present in the list


```python
nums = [22,11,55,44,33,66,88,77]
x = int(input('Enter the number : '))

for num in nums:
    if num == x:
        print(x, 'is present in the list')
```

## QUESTIONS ON BREAK

```python
nums = [22,11,55,44,33,66,88,77]
x = int(input('Enter the number : '))
iterations=0
for num in nums:
    iterations+=1
    if num == x:
        print(x,"is present in the list")
print("Number of iterations : ",iterations)
```

```python
nums = [22,11,55,44,33,66,88,77]
x = int(input('Enter the number : '))
iterations=0
for num in nums:
    iterations+=1
    if num == x:
        print(x,"is present in the list")
        break
    print(num)
print("Number of iterations : ",iterations)
```

###  Given a list of numbers take a number as input from the user and check whether the number is present in the list or not


```python
nums = [22,11,55,44,33,66,88,77,10,20,30,40,50,60,70,80,90,15,25,35,45,55,65,75,85,95]

x = int(input('Enter the number : '))

flag = 0
for num in nums:
    if num == x:
        flag = 1
        break

#print(flag)
if flag == 1:
    print(x,'is present in the list')
else:
    print(x,'is not present in the list')
```

### create a program that takes a number as input and prints whether it is prime or not


```python
num = int(input('Enter the number : '))

if num < 2:
    print(num,'is not a prime number')
else:
    for i in range(2, num):
        if num % i == 0:
            print(num, 'is not a prime number')
            break
    else:
        print(num, 'is a prime number')
```
