## Why We use Numpy?

- In Python we have lists that serve the purpose of arrays, but they are slow to process.

- NumPy aims to provide an array object that is up to 50x faster than traditional Python lists.

- The array object in NumPy is called ndarray, it provides a lot of supporting functions (statisitcal functions) that make working with ndarray very easy.

- Arrays are very frequently used in data science, where speed and resources are very important.

- you can do much more arithmetic operations with ndarrays that you cannot do with lists.Also operations with ndarrays will be much more faster than lists.

## Why is NumPy Faster Than Lists?
- NumPy arrays are stored at one continuous place in memory unlike lists, so processes can access and manipulate them very efficiently.

- This behavior is called locality of reference in computer science.

- This is the main reason why NumPy is faster than lists. Also it is optimized to work with latest CPU architectures.

- Lists is like storage data structure,these are not meant to perform operations

```python
#first we have to import library to use it
import numpy as np
```

```python
import time

# Create Python lists
list1 = list(range(1000000))
list2 = list(range(1000000))

# Create NumPy arrays
arr1 = np.array(list1)
arr2 = np.array(list2)

# Time list addition (element-wise using loop)
start = time.time()
list_result = [x + y for x, y in zip(list1, list2)]
end = time.time()
print("List addition time:", end - start)

# Time NumPy array addition (vectorized)
start = time.time()
array_result = arr1 + arr2
end = time.time()
print("NumPy addition time:", end - start)
```

```python
# NumPy arrays use less memory compared to Python lists for storing numerical data:
import sys
import numpy as np

# Create a list of 1000 numbers (from 0 to 999)
py_list = list(range(1000))

# Create a NumPy array of 1000 numbers (from /0 to 999)
np_array = np.arange(1000)

# Find memory used by Python list:
# sys.getsizeof(py_list) gives memory of the list object itself
# sum(sys.getsizeof(item) for item in py_list) adds memory of each number inside the list
list_memory = sys.getsizeof(py_list) + sum(sys.getsizeof(item) for item in py_list) #total memory = memory of the list + memory of each element inside it



# Find memory used by NumPy array (in bytes)
# np_array.nbytes gives total memory used by all numbers in the array
array_memory = np_array.nbytes

# Print both memory sizes
print("Memory used by Python list:", list_memory, "bytes")
print("Memory used by NumPy arra:", array_memory, "bytes")
```

```python
import sys
import numpy as np

# Create Python list of 1000 float numbers
py_list = [float(i) for i in range(1000)]

# Create NumPy array of 1000 float numbers
np_array = np.arange(1000, dtype='float64')

# Calculate memory used by Python list
list_memory = sys.getsizeof(py_list) + sum(sys.getsizeof(item) for item in py_list)

# Calculate memory used by NumPy array
array_memory = np_array.nbytes

# Print results
print("Memory used by Python float list:", list_memory, "bytes")
print("Memory used by NumPy float array:", array_memory, "bytes")
```

#### RESHAPE OF ARRAY
- Reshaping an array means changing its shape (i.e., dimensions) without changing its data.
- syntax : array.reshape(new_shape)
- new_shape must contain the same total number of elements as the original.
- reshape() returns a new array (doesn’t modify original).








```python
arr = np.array([1, 2, 3, 4, 5, 6])
reshaped = arr.reshape(2, 3)
print("Previous shape :",arr.shape)
print("Previous Dimension :",arr.ndim)
print()
print(reshaped)
print("New shape :",reshaped.shape)
print("New Dimension :",reshaped.ndim)
```

```python
# program to  check how many 2d possible shapes are possible
def possible_2d_reshapes(arr):
    size = arr.size          # Get total number of elements in the array
    reshapes = []            # Create an empty list to store valid shapes
    
    # Loop through all numbers from 1 to size
    for i in range(1, size + 1):
        if size % i == 0:    # Check if 'i' is a factor of size (i.e., divides evenly)
            # If yes, then (i, size//i) is a valid shape
            reshapes.append((i, size // i))
    
    return reshapes          # Return the list of valid 2D reshapes
arr=np.arange(20)
possible_2d_reshapes(arr)
```

- What does (i, size // i) mean?
- It means:Reshape the array into i rows, and size // i columns
- Because:
  - rows × columns = total elements
- - i × (size // i) = size ✅

#### TRANSPOSE OF AN NDARRAY

- Syntax - np.transpose(array, axes=None) OR array.T

```python
arr = np.array([[1, 2], [3, 4]])
print(arr.T)  
```

```python
arr = np.array([[1, 2], [3, 4]])
print(np.transpose(arr))
```

## Conditional Indexing
- Conditional Indexing in NumPy allows you to select elements from an array based on a condition. Instead of loops, you write a condition directly on the array.

- It's a very fast, clean, and Pythonic way to filter, modify, or analyze arrays.

```python
a = np.random.randint(11,100,(8,8))
a
```

```python
a > 50
```

```python
a[a > 50] #this is called boolean masking
```

Wherever the mask is True, the corresponding value from the original array is selected.

Wherever it is False, it is ignored.



```python
a[a <= 40]
```

```python
a[a == 16]
```

```python
b = a[a < 40]  # first this thing will be checked
print(b)
b[b > 20] 
```

### Slicing in Multidimensional Arrays
- Slicing is used to extract a portion of an array by specifying start, stop, and step indices.
For multidimensional arrays, slicing is done separately for each axis (dimension).

```python
a[:, 1]    # All rows, second column
```

```python
a[1,:]
```

```python
a[:5,:4]
```

```python
a[:8, :7]
```
