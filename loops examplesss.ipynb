```python
#Example 3: Using both break and continue
attempts = 0
while True:
    attempts += 1
    password = input("Enter password: ")
    
    if password == "":
        print("Empty input, try again.")
        continue  # skip empty input
    
    if password == "openAI123":
        print("Access granted.")
        break  # correct password, exit loop
    
    print("Wrong password. Try again.")
    
    if attempts >= 3:
        print("Too many attempts. Exiting...")
        break
```

```python
# Real-world Use: Search with break/else
numbers = [10, 20, 30, 40, 50]
target = 20
i = 0

while i < len(numbers):
    print(numbers[i])
    if numbers[i] == target:
        
        print(f"Found {target} at index {i}")
        break
    i += 1
else:
    print(f"{target} not found in the list.")
```

```python
nums = [11, 22, 33, 11, 33, 22, 11, 11, 33, 11]
print(len(nums))
freq = {}
i = 0
while i < len(nums):
    num = nums[i]
    print(num)
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1
    i += 1

print(freq)
```
