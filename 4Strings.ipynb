#### SLICING:ACCESSING A PART OF STRING:

-  step: how many steps to take, (can be negative for reverse):

```python

a="Hello World"
print(a[1:6])
```

####  NEGATIVE SLICING:
##### Negative slicing lets you slice strings from the end using negative indices — it combines the power of slicing and reverse indexing.
##### string[start:stop:step]
-  start: where to begin (inclusive)

-  stop: where to end (exclusive)

-  step: how many steps to take,(can be negative for reverse)


```python
s = "PYTHON"

print(s[-4:-1])   # "THO"  → from T to O (excludes -1)
print(s[-3:])     # "HON"  → from H to end
print(s[:-3])     # "PYT"  → from start to before H
print(s[-1:-4:-1])# "NOH"  → reverse slice from N to H

```

```python
s = "PROGRAMMING"
print(s[-8:-2:1])
print(s[-2:-8:-2])  #NMA
```

```python
#ISNUMERIC:
#The .isnumeric() method checks if all characters in the string are numeric characters, including:
#Digits (0–9)
#Unicode numeric characters (like superscripts, fractions, Roman numerals, etc.)
s = '123'
s.isnumeric() 
```

```python
print("123".isnumeric())        # True
print("²".isnumeric())          # True (superscript 2)
print("⅓".isnumeric())          # True (Unicode fraction)
print("123abc".isnumeric())     # False
print("".isnumeric())           # False
print("١٢٣".isnumeric())        # True (Arabic-Indic digits)
print("1.23".isnumeric())        # False
```

```python
#ISALPNUM:The .isalnum() method checks if a string contains only alphanumeric characters — that is:
#✅ Letters (a–z, A–Z)
#✅ Digits (0–9)
#❌ No spaces, punctuation, or symbols
s = '123abc'
s.isalnum()
```
